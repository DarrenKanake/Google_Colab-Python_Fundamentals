5. Handle User Input & Interaction
- Ensure error handling for invalid inputs.
- Guide the user with clear prompts.

```python
# --- Function to perform math operations ---
def math_operations(x, y):

    # Show the numbers
    print("The value of x =", x)
    print("The value of y =", y)

    # Math operations
    print("Addition: x + y =", x + y)
    print("Subtraction: x - y =", x - y)
    print("Multiplication: x * y =", x * y)

    # Division (check for zero first)
    if y == 0:
        print("Division: x / y = Cannot divide by zero!")
    else:
        print("Division: x / y =", x / y)


# --- Main loop ---
while True:

    # Get both numbers inside one try/except
    try:
        x = float(input("Enter the value of x (numbers only): "))
        y = float(input("Enter the value of y (numbers only): "))
    except ValueError:
        print("Not a number. Please try again.")
        continue

    # Call the function
    math_operations(x, y)

    # Ask if the user wants to continue
    again = input("\nDo you want to calculate again? (yes / no): ")

    if again == "no":
        print("Goodbye!")
        break
```

- Use formatted strings for better readability.

```python
# Function to perform math operations
def math_operations(x, y):

    # Show the numbers
    print("The value of x =", x)
    print("The value of y =", y)

    # Math operations
    print(f"Addition: x + y = {x + y:.2f}")
    print(f"Subtraction: x - y = {x - y:.2f}")
    print(f"Multiplication: x * y = {x * y:.2f}")

    # Division (check for zero first)
    if y == 0:
        print("Division: x / y = Cannot divide by zero!")
    else:
        print(f"Division: x / y = {x / y:.2f}")


# --- Main loop ---
while True:

    # Get both numbers inside one try/except
    try:
        x = float(input("Enter the value of x (numbers only): "))
        y = float(input("Enter the value of y (numbers only): "))
    except ValueError:
        print("Not a number. Please try again.")
        continue

    # Call the function
    math_operations(x, y)

    # Ask if the user wants to continue
    again = input("\nDo you want to calculate again? (yes / no): ")

    if again == "no":
        print("Goodbye!")
        break
```

```python
import math
# Function to perform math operations
def math_operations(x, y):

    # Show the numbers
    print("The value of x =", x)
    print("The value of y =", y)

    # Math operations
    print(f"Addition: x + y = {x + y:.2f}")
    print(f"Subtraction: x - y = {x - y:.2f}")
    print(f"Multiplication: x * y = {x * y:.2f}")
    print(f"Exponentiation: x ** y = {math.pow(x, y):.2f}")

    # Division (check for zero first)
    if y == 0:
        print("Division: x / y = Cannot divide by zero!")
    else:
        print(f"Division: x / y = {x / y:.2f}")
    if x < 0:
        print("Square root: Cannot calculate square root of a negative number!")
    else:
        print(f"Square root of x = {math.sqrt(x):.2f}")

# --- Main loop ---
while True:

    # Get both numbers inside one try/except
    try:
        x = float(input("Enter the value of x (numbers only): "))
        y = float(input("Enter the value of y (numbers only): "))
    except ValueError:
        print("Not a number. Please try again.")
        continue

    # Call the function
    math_operations(x, y)

    # Ask if the user wants to continue
    again = input("\nDo you want to calculate again? (yes / no): ")

    if again == "no":
        print("Goodbye!")
        break
```
